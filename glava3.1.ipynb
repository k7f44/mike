## Задание 1.1

```python
import matplotlib.pyplot as plt 
import numpy as np 
x_data = np.linspace(0, 20, 100) 
y_data = np.sin(x_data) 

fig = plt.figure() 
ax = plt.axes() 
ax.plot(x_data, y_data) 

ax.grid( linewidth=0.8,
        linestyle='--')

plt.plot(x_data, y_data, linewidth ='1', color='r')

plt.xlabel('x_data') 
plt.ylabel('y_data') 
plt.title('function sin(x)')
plt.legend(['sin(x)'])
plt.show() 
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
x_data = np.linspace(0, 20, 100) 
y_data = np.cos(x_data) 

fig = plt.figure() 
ax = plt.axes() 


ax.grid( linewidth=0.8,
        linestyle='--')

s=[1]
plt.scatter(x_data, y_data,marker= 's' , color='b')

plt.xlabel('x_data') 
plt.ylabel('y_data') 
plt.title('function cos(x)')
plt.legend(['cos(x)'])
plt.show() 
```
